```python
def convert_to_wide(csv_file_name, cust_id_col, col_to_wide):
    df = pd.read_csv(csv_file_name)
    df['customer_index'] = df.groupby(cust_id_col).cumcount() + 1
    df_new = df.pivot(index=cust_id_col, columns='customer_index', values=col_to_wide).reset_index()
    df_new.columns = [col[0] + "_" + str(col[1]) for col in df_new.columns]
    return df_new
```
